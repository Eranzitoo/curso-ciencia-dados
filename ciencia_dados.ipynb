```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Definir o estilo visual
sns.set(style="whitegrid")

# 2. Criar o gráfico de contagem (0 = Pagou em dia, 1 = Inadimplente)
plt.figure(figsize=(8, 5))

#/tmp/ipykernel_4596/1744049546.py:9: FutureWarning: 
#Passing `palette` without assigning `hue` is deprecated and will be removed in v0.14.0. Assign the `x` variable to `hue` and set `legend=False` for the same effect.
#sns.countplot(x='default payment next month', data=df, palette='viridis', legend=False)

# Atribuindo x ao hue e definindo legend=False como sugerido pelo aviso
sns.countplot(x='default payment next month', data=df, hue='default payment next month', palette='viridis', legend=False)

# 3. Adicionar títulos e legendas
plt.title('Proporção de Inadimplência no Próximo Mês')
plt.xlabel('Status de Inadimplência (0 = Não, 1 = Sim)')
plt.ylabel('Quantidade de Clientes')

# Mostrar o gráfico
plt.show()
```

🕵️ O que observar nesse gráfico:

    Pico da Idade: Onde a barra é mais alta?

    Outliers: Existem clientes com mais de 80 anos? Isso pode ser um dado real ou um erro de digitação.

    O gráfico "cai" rapidamente após os 50 anos? Isso indica que a empresa foca em um público mais jovem/adulto.

```python
# Criando um histograma com linha de densidade (KDE)
plt.figure(figsize=(10, 6))
sns.histplot(df['AGE'], bins=30, kde=True, color='skyblue')

# Personalizando o gráfico
plt.title('Distribuição de Idade dos Titulares de Conta')
plt.xlabel('Idade (Anos)')
plt.ylabel('Frequência (Quantidade de Clientes)')

plt.show()
```
